```python
#Using Otsu Thresholding in order to make the image black and white to be more legible.

import cv2
import numpy as np
from matplotlib import pyplot as plt
 
img = cv2.imread('warped2.jpg',0)
 
ret, imgf = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
 
plt.figure(figsize=(40,40))
plt.subplot(3,1,1), plt.imshow(img,cmap = 'gray')
plt.figure(figsize=(40,40))
plt.subplot(3,1,2), plt.imshow(imgf,cmap = 'gray')
plt.show()
```
